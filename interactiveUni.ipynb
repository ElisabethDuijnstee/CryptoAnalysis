```python
import numpy as np
import matplotlib.pyplot as plt

investment_usdc = 4000
spot = 1500

# Define the range of values for x
x = np.linspace(0, 5000, 100)

# Define the range of values for n
n_range = [1.05, 1.15, np]

# Define the colors for the lines
colors = ['#8367c7', '#b2a2cf', '#cab2d6']

# Create a figure and axis object
fig, ax = plt.subplots()

# Loop over the range of values for n and plot the PnL values against the price of Ethereum for each n
for i, n in enumerate(n_range):
    lower = spot / n
    upper = spot * n
    spot_range = max(min(spot, upper), lower)
    price_range = np.clip(x, lower, upper)
    y = investment_usdc*((x/np.sqrt(price_range)) - (x/np.sqrt(upper)) + np.sqrt(price_range) - np.sqrt(lower))/((spot/(np.sqrt(spot))) - (spot/np.sqrt(upper)) + np.sqrt(spot)-np.sqrt(lower)) -1 
    label = f'Lower: {lower:.2f}, Upper: {upper:.2f}'
    
    # Select the color based on the index
    color = colors[i]
    
    # Plot the line with the selected color
    ax.plot(x, y, label=label, color=color)
    ax.fill_between(x, y, 0, where=(lower < x) & (x < upper), color=color, alpha=0.5)
    

# Plot the vertical dashed line at the spot price
ax.axvline(x=spot, linestyle='--', color='black', label='Spot')

# Set the x and y axis labels
ax.set_xlabel('Price of Ethereum ($)')
ax.set_ylabel('Asset value (USDC)')

# Set the x and y axis limits
ax.set_xlim(380, 3300)
ax.set_ylim(1000, 6000)

# Add legend and grid
ax.legend()
ax.grid()

# Show the plot
plt.show()
```
